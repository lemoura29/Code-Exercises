```python
import pandas as pd
from tabulate import tabulate
import numpy as np
import seaborn as sns
import matplotlib.dates as mdates4
from matplotlib import pyplot as plt
from matplotlib.ticker import FormatStrFormatter
```

```python
xl = pd.ExcelFile('Data/Transactions.xlsx');
```

```python
df = xl.parse('Credit-Card-Transactions')
```

```python
df[['City','Country']]=df.City.str.split(',',expand=True)
```

```python
df['Year'] = df['Date'].dt.year
df['Quarter'] = df['Date'].dt.to_period('Q')
```

```python
dfCity = df.groupby(["Year", "City"])["Amount"].sum().reset_index()
```

<h2>TOP-5-Cities-Expenses<h2>

```python
df_graph = pd.DataFrame()
for i in dfCity['Year'].unique():
    df1 = dfCity[dfCity['Year']==i].sort_values(by=['Amount','City'],ascending=False ).head(5)
    df_graph = pd.concat([df_graph, df1])
    df1 = df1.pivot_table('Amount', index='Year', columns='City', sort= False).sort_values(by=i, ascending=False, axis=1)
    
   
    print(tabulate(df1, headers = 'keys', tablefmt = 'fancy_grid', numalign="right",  disable_numparse=True,floatfmt=".4f"))
    

    
```

```python
sns.set(style='darkgrid')


g = sns.catplot(x="Year", y="Amount", hue="City",
                data=df_graph, kind="bar",
                height=10,aspect=.6,  palette="gist_earth", legend=False)


g.set_axis_labels('City', 'Amount in Million(CAD)', fontsize=18)
plt.legend(loc='upper right', title = 'Expenses Type', fontsize=12, ncol=3)

g.fig.set_size_inches(15, 8)
g.fig.suptitle('TOP3-Expenses-Analysis', fontsize=20)


ax = g.facet_axis(0, 0)
ax.yaxis.set_major_formatter(FormatStrFormatter('%d'))

for c in ax.containers:
    labels = [f'{(v.get_height() / 1000000):.1f}M' for v in c]
    ax.bar_label(c, labels=labels, label_type='edge', fontsize=9)
    

plt.show()

```

<h2>Gender-Expenses-Analysis<h2>

```python
dfQuarter = df.groupby(["Year", "Quarter",'Gender'])["Amount"].sum().reset_index()
```

```python
year = dfQuarter['Year'].unique()
for i in year:
    df1 = dfQuarter[dfQuarter['Year']==i].pivot_table('Amount', index=["Year","Gender"], columns=['Quarter'], sort= False).reset_index()
    print(tabulate(df1, headers = 'keys', tablefmt = 'fancy_grid', numalign="right",  disable_numparse=True, showindex=False))
```

```python

q = sns.catplot(data=dfQuarter, x="Quarter", y="Amount", hue="Gender", 
                markers=["^", "X"], kind='point', legend=False, palette="PRGn")
q.set_axis_labels('Quarters', 'Amount in Million(CAD)', fontsize=18)
q.fig.set_size_inches(15, 8)
q.fig.suptitle('Gender-Expenses-Analysis', fontsize=20)
plt.legend(loc='upper right', title = 'Gender', fontsize=12, ncol=3)


plt.show()
```

<h2>TOP3-Expenses-Analysis<h2>

```python
dfExpenses = df.groupby(["Year", "Exp Type"])["Amount"].sum().reset_index()
```

```python
year = dfExpenses['Year'].unique()
df_graph = pd.DataFrame()
for i in year:
    df1 = dfExpenses[dfExpenses['Year']==i].sort_values(by=['Amount','Exp Type'],ascending=False).head(3)
    df_graph = pd.concat([df_graph, df1])
    df1 = df1.pivot_table('Amount', index="Year", columns=['Exp Type']).sort_values(by=i, axis=1)
    
    print(tabulate(df1, headers = 'keys', tablefmt = 'fancy_grid', numalign="right",  disable_numparse=True))
```

```python
sns.set(style='darkgrid')


g = sns.catplot(x="Year", y="Amount", hue="Exp Type",
                data=df_graph, kind="bar",
                height=10,aspect=.6,  palette="ocean_r", legend=False)


g.set_axis_labels('Expenses Type', 'Amount in Million(CAD)', fontsize=18)
plt.legend(loc='upper right', title = 'Expenses Type', fontsize=12, ncol=3)

g.fig.set_size_inches(15, 8)
g.fig.suptitle('TOP3-Expenses-Analysis', fontsize=20)


ax = g.facet_axis(0, 0)
ax.yaxis.set_major_formatter(FormatStrFormatter('%d'))

for c in ax.containers:
    labels = [f'{(v.get_height() / 1000000):.1f}M' for v in c]
    ax.bar_label(c, labels=labels, label_type='edge')


plt.show()

```
